**Scenario 1**

 Description: Analyze a sample e-commerce dataset to understand product categories, sales trends, and missing information.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/data.csv", encoding="latin1")
print("Priyadharshini B - 24BAD093")
print("Dataset loaded successfully!\n")

# First 5 rows
print("First 5 rows of the dataset:")
print(df.head(), "\n")

# Last 5 rows
print("Last 5 rows of the dataset:")
print(df.tail(), "\n")

# Dataset information
print("Dataset Information:")
df.info()
print("\n")

# Statistical summary
print("Statistical Description:")
print(df.describe())
print("\n")

print("Missing values in each column:")
print(df.isnull().sum())
print("\n")

top_products = (
    df.groupby("Description")["Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure()
top_products.plot(kind="bar")
plt.title("Top 10 Products by Sales Quantity")
plt.xlabel("Product")
plt.ylabel("Total Quantity Sold")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Line chart
plt.figure()
df["Quantity"].head(50).plot(kind="line")
plt.title("Sales Trend (First 50 Records)")
plt.xlabel("Record Index")
plt.ylabel("Quantity Sold")
plt.show()
print("IN-LAB TASK COMPLETED SUCCESSFULLY")
```

**Scenario 2**

Description: Examine a patient dataset to identify missing health metrics and visualize distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/diabetes.csv")
print("Priyadharshini B - 24BAD093")
print("Dataset loaded successfully!\n")

print("First 5 rows:")
print(df.head(), "\n")

print("Last 5 rows:")
print(df.tail(), "\n")

print("Dataset Information:")
df.info()
print("\n")

print("Statistical Description:")
print(df.describe())
print("\n")
columns_with_missing = [
    "Glucose", "BloodPressure", "SkinThickness",
    "Insulin", "BMI"
]

df[columns_with_missing] = df[columns_with_missing].replace(0, pd.NA)

print("Missing values in each column:")
print(df.isnull().sum())
print("\n")
plt.figure()
df["Glucose"].dropna().plot(kind="hist", bins=20)
plt.title("Distribution of Glucose Levels")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()

# Boxplot: Glucose levels
plt.figure()
plt.boxplot(df["Glucose"].dropna())
plt.title("Boxplot of Glucose Levels")
plt.ylabel("Glucose Level")
plt.show()

# Histogram: Age distribution
plt.figure()
df["Age"].plot(kind="hist", bins=20)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Boxplot: Age distribution
plt.figure()
plt.boxplot(df["Age"])
plt.title("Boxplot of Age Distribution")
plt.ylabel("Age")
plt.show()

print("Pattern Observations:")
print("- Higher glucose values are commonly observed in diabetic patients.")
print("- Age distribution shows majority of patients are middle-aged.")
print("- Presence of outliers in glucose and BMI indicates high-risk patients.")

print("\nSCENARIO 2 COMPLETED SUCCESSFULLY")
```

**Scenario 3**

Description: Investigate a housing dataset for missing features and relationships between house attributes.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("/content/Housing.csv")
print("Priyadharshini B - 24BAD093")
print("Dataset loaded successfully!\n")

print("Column names:")
print(df.columns, "\n")

print("First 5 rows:")
print(df.head(), "\n")

print("Dataset Information:")
df.info()
print("\n")

print("Statistical Description:")
print(df.describe())
print("\n")

print("Missing values in each column:")
print(df.isnull().sum())
print("\n")

# Scatter plot: Area vs Price
plt.figure()
plt.scatter(df["area"], df["price"])
plt.title("House Area vs Price")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price")
plt.show()

# Scatter plot: Bedrooms vs Price
plt.figure()
plt.scatter(df["bedrooms"], df["price"])
plt.title("Bedrooms vs Price")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Price")
plt.show()

# Select only numerical columns
numerical_df = df.select_dtypes(include=[np.number])

# Compute correlation matrix
correlation_matrix = numerical_df.corr()

plt.figure()
plt.imshow(correlation_matrix, cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title("Correlation Heatmap of Housing Features")
plt.tight_layout()
plt.show()

print("Trend Interpretation:")
print("- House price increases as area increases.")
print("- Number of bedrooms shows moderate correlation with price.")
print("- Strongly correlated features are important for price prediction.")
print("- Weak correlations indicate less influence on house price.")

print("\nSCENARIO 3 COMPLETED SUCCESSFULLY")
```

**Scenario 4**

Description: Analyze banking data to detect missing account information and understand customer demographics.

```python
import pandas as pd
import matplotlib.pyplot as plt

print("Priyadharshini B - 24BAD093")

# Load dataset
df = pd.read_csv("/content/marketing_campaign.csv", sep="\t")

print("\nFirst 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

df["Age"] = 2024 - df["Year_Birth"]

df["Total_Spending"] = (
    df["MntWines"] +
    df["MntFruits"] +
    df["MntMeatProducts"] +
    df["MntFishProducts"] +
    df["MntSweetProducts"] +
    df["MntGoldProds"]
)
plt.hist(df["Age"], bins=20)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()

# INCOME BOXPLOT
plt.boxplot(df["Income"].dropna(), vert=False)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.show()

# SPENDING BOXPLOT
plt.boxplot(df["Total_Spending"], vert=False)
plt.title("Customer Spending Pattern")
plt.xlabel("Total Spending")
plt.show()

print("\nINSIGHTS:")
print("1. Most customers belong to middle age group.")
print("2. Income contains missing values and outliers.")
print("3. Customers with higher income tend to spend more.")
```
